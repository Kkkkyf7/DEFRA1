```python
import pandas as pd

df = pd.read_csv(r"C:\Users\qp24297\OneDrive - University of Bristol\Final_dataset\merged_dataset_with_year.csv")
df['SUBSTANCE NAME'] = df['SUBSTANCE NAME'].str.title()

substances = ["Indeno(1,2,3-Cd)Pyrene", "Benzo(A)Pyrene", "Cypermethrin", "Fluoranthene", "Copper"]

results = []

for substance in substances:
    df_sub = df[df['SUBSTANCE NAME'] == substance]
    
    total_quantity = df_sub['QUANTITY RELEASED (KG)'].sum()
    mean_quantity = df_sub['QUANTITY RELEASED (KG)'].mean()
    std_quantity = df_sub['QUANTITY RELEASED (KG)'].std()
    min_quantity = df_sub['QUANTITY RELEASED (KG)'].min()
    max_quantity = df_sub['QUANTITY RELEASED (KG)'].max()
    mean_reporting_threshold = df_sub['REPORTING THRESHOLD (KG)'].mean()
    num_observations = df_sub.shape[0]
    percent_below_threshold = (df_sub['QUANTITY RELEASED (KG)'] < df_sub['REPORTING THRESHOLD (KG)']).mean() * 100
    
    num_sites_all = df_sub['SITE ADDRESS'].nunique()
    site_years_all = df_sub.groupby('SITE ADDRESS')['Year'].nunique()
    mean_years_per_site_all = site_years_all.mean()
    
    results.append({
        "Substance": substance,
        "Total Quantity Released (kg)": total_quantity,
        "Mean": mean_quantity,
        "Std Dev": std_quantity,
        "Min": min_quantity,
        "Max": max_quantity,
        "Mean Reporting Threshold (kg)": mean_reporting_threshold,
        "Number of Observations": num_observations,
        "Percent < Threshold": percent_below_threshold,
        "Nb of Different Sites (all)": num_sites_all,
        "Mean Nb Years per Site (all)": mean_years_per_site_all
    })

df_results_all = pd.DataFrame(results)

pd.set_option("display.float_format", lambda x: f"{x:,.2f}")
print(df_results_all)
```
