```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

df = pd.read_csv("regdf_with_sams.csv")

def ensure_log(df, col, allow_zero=False):
    log_col = f"log_{col}"
    if log_col not in df.columns:
        if allow_zero:
            df[log_col] = np.log1p(df[col].clip(lower=0))
        else:
            df[log_col] = np.where(df[col] > 0, np.log(df[col]), np.nan)
    return log_col

log_Population           = ensure_log(df, "Population")
log_Active_enterprises   = ensure_log(df, "Active_enterprises")
log_Median_weekly_pay    = ensure_log(df, "Median_weekly_pay")
log_Pop_density          = ensure_log(df, "Population_density")
log_Ent_density          = ensure_log(df, "Enterprise_density")
log_Coastline_ha         = ensure_log(df, "Clipped_to_coastline_ha")
log_Inland_water_ha      = ensure_log(df, "Inland_water_ha", allow_zero=True)

YEARS = list(range(2016, 2024))
N_UTLA = 127
i_2021 = YEARS.index(2021)
start, end = i_2021 * N_UTLA, (i_2021 + 1) * N_UTLA
df_2021 = df.iloc[start:end].copy()

def run_model(cols, dep="log_Pollution_toxic"):
    X = df_2021[cols].copy()
    X = sm.add_constant(X)
    y = df_2021[dep]
    model = sm.OLS(y, X).fit(cov_type="HC3")  # robust SEs
    return model

M0_cols = [log_Population, log_Active_enterprises]
M1_cols = M0_cols + [log_Median_weekly_pay, "Unemployment_rate"]
M2_cols = M1_cols + [log_Coastline_ha, log_Inland_water_ha]
M3_cols = [log_Pop_density, log_Ent_density, log_Median_weekly_pay, "Unemployment_rate"]

m0 = run_model(M0_cols)
m1 = run_model(M1_cols)
m2 = run_model(M2_cols)
m3 = run_model(M3_cols)

print("\n=== M0: Size (counts only, logs) ===")
print(m0.summary())

print("\n=== M1: Size + income/labour (logs, unemp in levels) ===")
print(m1.summary())

print("\n=== M2: + geography (area in logs) ===")
print(m2.summary())

print("\n=== M3: Densities (logs) ===")
print(m3.summary())
```

```python
import os
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_white, het_breuschpagan

YEARS = list(range(2016, 2024))
N_UTLA = 127

if os.path.exists("regdf_with_sams.csv"):
    df = pd.read_csv("regdf_with_sams.csv")
else:
    df = pd.read_csv("regdf.csv")
    sams = pd.read_csv("SAMS2021.csv")
    area_repeat = pd.concat([sams] * len(YEARS), ignore_index=True)
    df["Clipped_to_coastline_ha"] = area_repeat["Clipped to the Coastline (Area in Hectares)"].values
    df["Inland_water_ha"] = area_repeat["Area of Inland Water (Hectares)"].values
    df.to_csv("regdf_with_sams.csv", index=False)
    print("Saved merged file: regdf_with_sams.csv")

df["log_Population"] = np.log(df["Population"])
df["log_Active_enterprises"] = np.log(df["Active_enterprises"])
df["log_Median_weekly_pay"] = np.log(df["Median_weekly_pay"])
df["log_Coastline_ha"] = np.log(df["Clipped_to_coastline_ha"])
df["log_Inland_water_ha"] = np.log1p(df["Inland_water_ha"])  # log1p handles zero safely

model_vars = [
    "log_Population",
    "log_Active_enterprises",
    "log_Median_weekly_pay",
    "Unemployment_rate",
    "log_Coastline_ha",
    "log_Inland_water_ha"
]
dep_var = "log_Pollution_toxic"
df.to_csv("regdf_with_sams.csv", index=False)
for i, year in enumerate(YEARS):
    start, end = i * N_UTLA, (i + 1) * N_UTLA
    d = df.iloc[start:end].copy()

    # drop NA
    d = d[[dep_var] + model_vars].dropna()
    if d.empty:
        print(f"\n=== {year} ===\nNo data after dropping NAs.")
        continue

    X = sm.add_constant(d[model_vars], has_constant="add")
    y = d[dep_var]

    mod = sm.OLS(y, X).fit()

    print(f"\n=== {year} ===")
    try:
        lm, lm_p, fval, f_p = het_white(mod.resid, mod.model.exog)
        print(f"White's test: LM={lm:.3f}, p={lm_p:.4f} | F={fval:.3f}, p={f_p:.4f}")
        if lm_p < 0.05 or f_p < 0.05:
            print("→ Evidence of heteroskedasticity (White). Use robust SEs (HC3) or re-specify.")
        else:
            print("→ No strong evidence of heteroskedasticity (White).")
    except AssertionError:
        bp = het_breuschpagan(mod.resid, mod.model.exog)
        lm, lm_p, fval, f_p = bp
        print(f"White's test failed due to collinearity; using Breusch–Pagan instead.")
        print(f"Breusch–Pagan: LM={lm:.3f}, p={lm_p:.4f} | F={fval:.3f}, p={f_p:.4f}")
        if lm_p < 0.05 or f_p < 0.05:
            print("→ Evidence of heteroskedasticity (BP). Use robust SEs (HC3) or re-specify.")
        else:
            print("→ No strong evidence of heteroskedasticity (BP).")
```

```python
import os
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_white, het_breuschpagan

YEARS = list(range(2016, 2024))  # 8 years
N_YEARS = len(YEARS)
N_UTLA = 127

df = pd.read_csv("regdf.csv")         
sams = pd.read_csv("SAMS2021.csv")     

assert len(sams) == N_UTLA, f"SAMS rows ({len(sams)}) != N_UTLA ({N_UTLA})"
assert len(df) == N_UTLA * N_YEARS, f"df rows ({len(df)}) != {N_UTLA*N_YEARS}"

coast_vec  = np.repeat(sams["Clipped to the Coastline (Area in Hectares)"].to_numpy(), N_YEARS)
inland_vec = np.repeat(sams["Area of Inland Water (Hectares)"].to_numpy(), N_YEARS)

df["Clipped_to_coastline_ha"] = coast_vec
df["Inland_water_ha"]         = inland_vec

for utla_idx in range(N_UTLA):
    s, e = utla_idx * N_YEARS, (utla_idx + 1) * N_YEARS
    assert df["Clipped_to_coastline_ha"].iloc[s:e].nunique() == 1, f"coastline not constant in UTLA {utla_idx}"
    assert df["Inland_water_ha"].iloc[s:e].nunique() == 1, f"inland water not constant in UTLA {utla_idx}"

df.to_csv("regdf_with_sams.csv", index=False)

def safe_log(x):
    x = pd.to_numeric(x, errors="coerce")
    x = x.where(x > 0)  # nonpositive -> NaN
    return np.log(x)

df["log_Population"]         = safe_log(df["Population"])
df["log_Active_enterprises"] = safe_log(df["Active_enterprises"])
df["log_Median_weekly_pay"]  = safe_log(df["Median_weekly_pay"])
df["log_Coastline_ha"]       = safe_log(df["Clipped_to_coastline_ha"])
df["log_Inland_water_ha"]    = np.log1p(pd.to_numeric(df["Inland_water_ha"], errors="coerce"))

df["Year"] = np.tile(YEARS, N_UTLA)

model_vars = [
    "log_Population",
    "log_Active_enterprises",
    "log_Median_weekly_pay",
    "Unemployment_rate",
    "log_Coastline_ha",
    "log_Inland_water_ha"
]
dep_var = "log_Pollution_toxic"

for year in YEARS:
    d = df[df["Year"] == year].copy()
    d = d[[dep_var] + model_vars].dropna()

    if d.empty:
        print(f"\n=== {year} ===\nNo data after dropping NAs.")
        continue

    X = sm.add_constant(d[model_vars], has_constant="add")
    y = d[dep_var]

    mod = sm.OLS(y, X).fit()

    print(f"\n=== {year} ===")
    try:
        lm, lm_p, fval, f_p = het_white(mod.resid, mod.model.exog)
        print(f"White's test: LM={lm:.3f}, p={lm_p:.4f} | F={fval:.3f}, p={f_p:.4f}")
        if lm_p < 0.05 or f_p < 0.05:
            print("→ Evidence of heteroskedasticity (White). Use robust SEs (HC3) or re-specify.")
        else:
            print("→ No strong evidence of heteroskedasticity (White).")
    except AssertionError:
        lm, lm_p, fval, f_p = het_breuschpagan(mod.resid, mod.model.exog)
        print("White's test failed due to collinearity; using Breusch–Pagan instead.")
        print(f"Breusch–Pagan: LM={lm:.3f}, p={lm_p:.4f} | F={fval:.3f}, p={f_p:.4f}")
        if lm_p < 0.05 or f_p < 0.05:
            print("→ Evidence of heteroskedasticity (BP). Use robust SEs (HC3) or re-specify.")
        else:
            print("→ No strong evidence of heteroskedasticity (BP).")
```

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np

alpha = 0.10  # cutoff for squared terms


n_utlas = 127
n_years = 8
years = list(range(2016, 2016 + n_years))

df["Year"] = np.tile(years, n_utlas)

linear_vars = [
    "log_Population",
    "log_Active_enterprises",
    "log_Median_weekly_pay",
    "Unemployment_rate",
    "log_Coastline_ha",
    "log_Inland_water_ha"
]

squared_vars = [
    "log_Population_sq",
    "log_Active_enterprises_sq",
    "log_Median_weekly_pay_sq",
    "Unemployment_rate_sq",
    "log_Coastline_ha_sq"
]


for year in years:
    df_year = df[df["Year"] == year].copy().dropna(subset=[dep_var] + linear_vars + squared_vars)
    y = df_year[dep_var]

    # --- Start with all linear vars + all squared vars ---
    keep_squares = squared_vars.copy()

    while True:
        final_vars = linear_vars + keep_squares
        X_final = sm.add_constant(df_year[final_vars])

        model = sm.OLS(y, X_final).fit(cov_type="HC3")

        # Keep only squares significant at alpha
        new_keep = [v for v in keep_squares if model.pvalues.get(v, 1) < alpha]

        # Stop if the set doesn’t change
        if set(new_keep) == set(keep_squares):
            break
        keep_squares = new_keep

    final_vars = linear_vars + keep_squares
    X_final = sm.add_constant(df_year[final_vars])
    model_final = sm.OLS(y, X_final).fit(cov_type="HC3")

    print(f"\n=== {year} ===")
    print("Variables kept:", ["const"] + final_vars)
    print(model_final.summary())
```

```python
import os
import pandas as pd
import numpy as np
import statsmodels.api as sm

YEARS   = list(range(2016, 2024))  
N_YEARS = len(YEARS)
N_UTLA  = 127
alpha   = 0.10                     
dep_var = "log_Pollution_toxic"    

df   = pd.read_csv("regdf.csv")          
sams = pd.read_csv("SAMS2021.csv")     

assert len(sams) == N_UTLA, f"SAMS rows ({len(sams)}) != N_UTLA ({N_UTLA})"
assert len(df)   == N_UTLA * N_YEARS, f"df rows ({len(df)}) != {N_UTLA*N_YEARS}"

df["Clipped_to_coastline_ha"] = np.repeat(
    sams["Clipped to the Coastline (Area in Hectares)"].to_numpy(), N_YEARS
)
df["Inland_water_ha"] = np.repeat(
    sams["Area of Inland Water (Hectares)"].to_numpy(), N_YEARS
)

df["Year"] = np.tile(YEARS, N_UTLA)

def safe_log(x):
    x = pd.to_numeric(x, errors="coerce")
    x = x.where(x > 0)  # nonpositive -> NaN
    return np.log(x)

df["log_Population"]         = safe_log(df["Population"])
df["log_Active_enterprises"] = safe_log(df["Active_enterprises"])
df["log_Median_weekly_pay"]  = safe_log(df["Median_weekly_pay"])
df["log_Coastline_ha"]       = safe_log(df["Clipped_to_coastline_ha"])
df["log_Inland_water_ha"]    = np.log1p(pd.to_numeric(df["Inland_water_ha"], errors="coerce"))

df["log_Population_sq"]            = df["log_Population"]**2
df["log_Active_enterprises_sq"]    = df["log_Active_enterprises"]**2
df["log_Median_weekly_pay_sq"]     = df["log_Median_weekly_pay"]**2
df["Unemployment_rate_sq"]         = df["Unemployment_rate"]**2   # squared in levels
df["log_Coastline_ha_sq"]          = df["log_Coastline_ha"]**2

linear_vars = [
    "log_Population",
    "log_Active_enterprises",
    "log_Median_weekly_pay",
    "Unemployment_rate",
    "log_Coastline_ha",
    "log_Inland_water_ha",
]

squared_vars = [
    "log_Population_sq",
    "log_Active_enterprises_sq",
    "log_Median_weekly_pay_sq",
    "Unemployment_rate_sq",
    "log_Coastline_ha_sq",
]

for year in YEARS:
    df_year = df[df["Year"] == year].copy()

    needed_cols = [dep_var] + linear_vars + squared_vars
    df_year = df_year[needed_cols].replace([np.inf, -np.inf], np.nan).dropna()
    y = df_year[dep_var]

    if year <= 2017:
        cov_type = None
        se_label = "classical OLS"
    else:
        cov_type = "HC3"
        se_label = "robust HC3"

    keep_squares = squared_vars.copy()
    while True:
        final_vars = linear_vars + keep_squares
        X_final = sm.add_constant(df_year[final_vars], has_constant="add")
        model = sm.OLS(y, X_final).fit(cov_type=cov_type) if cov_type else sm.OLS(y, X_final).fit()

        new_keep = [v for v in keep_squares if model.pvalues.get(v, 1.0) < alpha]
        if set(new_keep) == set(keep_squares):
            break
        keep_squares = new_keep

    final_vars = linear_vars + keep_squares
    X_final = sm.add_constant(df_year[final_vars], has_constant="add")
    model_final = sm.OLS(y, X_final).fit(cov_type=cov_type) if cov_type else sm.OLS(y, X_final).fit()

    print(f"\n=== {year} === ({se_label})")
    print("Variables kept:", ["const"] + final_vars)
    print(model_final.summary())
```

```python
import itertools
import pandas as pd
import statsmodels.api as sm
from linearmodels.iv import IV2SLS

df = pd.read_csv("regdf_with_sams.csv")
YEARS = list(range(2016, 2024))
N_UTLA = 127
df["Year"] = YEARS * N_UTLA
df["UTLA_id"] = [i for i in range(N_UTLA)] * len(YEARS)

d2019 = df[df["Year"] == 2019].copy()

dep_var = "log_Pollution_toxic"
main_exogs = ["log_Population", "log_Active_enterprises", "log_Median_weekly_pay", "Unemployment_rate"]

candidate_instr = [
    "log_Coastline_ha",
    "log_Inland_water_ha",
]

d2019 = d2019.dropna(subset=[dep_var] + main_exogs + candidate_instr)

def first_stage_F(yvar, Xvars):
    X = sm.add_constant(d2019[Xvars])
    y = d2019[yvar]
    mod = sm.OLS(y, X).fit()
    return mod.fvalue, mod.f_pvalue

def wu_hausman_test(test_var, instruments):
    other_exogs = [v for v in main_exogs if v != test_var]
    formula = f"{dep_var} ~ 1 + {' + '.join(other_exogs)} + [{test_var} ~ {' + '.join(instruments)}]"
    model = IV2SLS.from_formula(formula, data=d2019).fit()
    return model.wu_hausman()

results = []
for var in main_exogs:  # suspected endogenous
    for L in range(1, len(candidate_instr) + 1):
        for combo in itertools.combinations(candidate_instr, L):
            f_stat, p_val = first_stage_F(var, list(combo))
            try:
                test = wu_hausman_test(var, list(combo))
                wu_stat, wu_p = test.stat, test.pval
            except Exception as e:
                wu_stat, wu_p = None, None
            results.append({
                "Endog_var": var,
                "Instruments": combo,
                "First_stage_F": f_stat,
                "First_stage_p": p_val,
                "Wu_stat": wu_stat,
                "Wu_p": wu_p,
                "Endogenous?": "Yes" if wu_p is not None and wu_p < 0.05 else "No/Unknown"
            })

res_df = pd.DataFrame(results).sort_values(by="First_stage_F", ascending=False)
print(res_df)
```

```python
import pandas as pd
import itertools
from linearmodels.iv import IV2SLS
import statsmodels.api as sm

def run_endogeneity_tests_per_year(df_by_year, dep_var='log_Pollution_toxic', f_threshold=10):
    results_all = {}
    
    for year, df in df_by_year.items():
        print(f"\n=== {year} ===")
        

        suspected_vars_map = {
            2016: ['log_Active_enterprises', 'Unemployment_rate'],
            2017: ['Unemployment_rate'],
            2018: ['log_Population', 'log_Active_enterprises'],
            2019: ['log_Active_enterprises', 'log_Median_weekly_pay', 'Unemployment_rate'],
            2020: ['Unemployment_rate'],
            2021: ['log_Population', 'Unemployment_rate'],
            2022: ['log_Population', 'log_Active_enterprises', 'log_Median_weekly_pay'],
            2023: ['log_Population', 'log_Median_weekly_pay', 'Unemployment_rate']
        }
        
        suspected_vars = suspected_vars_map.get(year, [])
        
     
        potential_instr = ['log_Coastline_ha', 'log_Inland_water_ha']
        
        year_results = []
        
        for endog_var in suspected_vars:
            # Step 1: First-stage regression for each instrument separately
            strong_instr = []
            for instr in potential_instr:
                y = df[endog_var]
                X = sm.add_constant(df[[instr]])
                model = sm.OLS(y, X).fit()
                f_stat = model.fvalue
                if f_stat is not None and f_stat > f_threshold:
                    strong_instr.append(instr)
            
            if not strong_instr:
                print(f"No strong instruments found for {endog_var} in {year}. Skipping.")
                continue
            

            for L in range(1, len(strong_instr) + 1):
                for combo in itertools.combinations(strong_instr, L):
                    exog_vars = [v for v in suspected_vars if v != endog_var]
                    formula = f"{dep_var} ~ 1 + [{' + '.join(exog_vars)}] + [{endog_var} ~ {' + '.join(combo)}]"
                    
                    try:
                        model = IV2SLS.from_formula(formula, data=df).fit()
                        wu = model.wu_hausman()
                        year_results.append({
                            'Endog_var': endog_var,
                            'Instruments': combo,
                            'Wu_stat': wu.stat,
                            'Wu_p': wu.pval,
                            'Endogenous?': 'Yes' if wu.pval < 0.05 else 'No/Unknown'
                        })
                    except Exception as e:
                        print(f"Error with {endog_var}, instruments {combo}: {e}")
        
        results_all[year] = pd.DataFrame(year_results)
    
    return results_all
```

```python

years = [2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]

rows_per_year = 127
df_by_year = {
    year: df.iloc[i*rows_per_year:(i+1)*rows_per_year].copy()
    for i, year in enumerate(years)
}

results_all = run_endogeneity_tests_per_year(df_by_year)

for year, res in results_all.items():
    print(f"\nResults for {year}:")
    if not res.empty:
        print(res)
    else:
        print("No valid strong instruments / tests run.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm

years = [2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
rows_per_year = 127

df_by_year = {
    year: df.iloc[i*rows_per_year:(i+1)*rows_per_year].copy()
    for i, year in enumerate(years)
}

vars_to_plot = [
    'log_Population',
    'log_Active_enterprises',
    'log_Median_weekly_pay',
    'Unemployment_rate',   # not logged in your model
    'log_Coastline_ha',
    'log_Inland_water_ha'
]

for var in vars_to_plot:
    for year, df_year in df_by_year.items():
        x = df_year[var]
        y = df_year['log_Pollution_toxic']
        
        # Fit linear model
        X_lin = sm.add_constant(x)
        lin_model = sm.OLS(y, X_lin).fit()
        y_pred_lin = lin_model.predict(X_lin)
        
        # Fit quadratic model
        X_quad = sm.add_constant(pd.DataFrame({var: x, f'{var}_sq': x**2}))
        quad_model = sm.OLS(y, X_quad).fit()
        y_pred_quad = quad_model.predict(X_quad)
        
        # Plot
        plt.figure(figsize=(6,4))
        plt.scatter(x, y, alpha=0.6, label='Data')
        plt.plot(x, y_pred_lin, color='red', label='Linear fit')
        plt.plot(x, y_pred_quad, color='green', linestyle='--', label='Quadratic fit')
        plt.xlabel(var)
        plt.ylabel('log_Pollution_toxic')
        plt.title(f'{year}: {var} vs log_Pollution_toxic')
        plt.legend()
        plt.tight_layout()
        plt.show()
```
