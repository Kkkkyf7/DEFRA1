```python
import pandas as pd
import matplotlib.pyplot as plt

# Read Excel file
file_path = "C:/Users/ZITIAO/Downloads/pollution_no_gas_cleaned.xlsx"
df = pd.read_excel(file_path)

# Check the unique value and frequency of column 'REGULATED INDUSTRY SUB SECTOR LABEL'
sub_sector_counts = df['REGULATED INDUSTRY SUB SECTOR LABEL'].value_counts().sort_index()

# Draw a histogram
plt.figure(figsize=(12, 6))
sub_sector_counts.plot(kind='bar', color='skyblue')

plt.xlabel('Regulated Industry Sub Sector Label')
plt.ylabel('Frequency')
plt.title('Histogram of REGULATED INDUSTRY SUB SECTOR LABEL')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
