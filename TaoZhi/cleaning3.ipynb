```python
import pandas as pd
import os

# Folder containing the Excel files (use '.' if they're in the current working directory)
folder_path = "."

# List of Excel filenames based on original DataFrame names
file_names = [
    "df_no_gas.xlsx", "df22_no_gas.xlsx", "df21_no_gas.xlsx", "df20_no_gas.xlsx",
    "df19_no_gas.xlsx", "df18_no_gas.xlsx", "df17_no_gas.xlsx", "df16_no_gas.xlsx"
]

# Dictionary to store loaded DataFrames
dfs2 = {}

# Read each Excel file into a DataFrame
for file in file_names:
    file_path = os.path.join(folder_path, file)
    df_name = file.replace(".xlsx", "")
    dfs2[df_name] = pd.read_excel(file_path)

# Optionally display the keys (i.e., DataFrame names)
print("Loaded DataFrames:")
print(list(dfs2.keys()))
```

```python
# Dictionary to track how many copper rows are found and added
copper_counts = {}

# Step 1: Store original lengths of dfs
original_lengths = {name: len(df) for name, df in dfs.items()}

for df2_name, df2 in dfs2.items():
    # Count copper rows
    copper_rows = df2[df2['SUBSTANCE NAME'].str.lower() == 'copper']
    copper_count = len(copper_rows)
    copper_counts[df2_name] = copper_count

    # Special case fix for 'df_no_gas'
    if df2_name == "df_no_gas":
        corresponding_name = "d23_v3"
    else:
        corresponding_name = df2_name.replace("df", "d").replace("_no_gas", "") + "_v3"

    # Step 2: Append copper rows
    if corresponding_name in dfs:
        dfs[corresponding_name] = pd.concat([dfs[corresponding_name], copper_rows], ignore_index=True)
    else:
        print(f"⚠️ Warning: {corresponding_name} not found in dfs")

# Step 3: Check that row counts increased correctly
print("\n✅ Validation Report:")
for df2_name, count_added in copper_counts.items():
    if df2_name == "df_no_gas":
        corresponding_name = "d23_v3"
    else:
        corresponding_name = df2_name.replace("df", "d").replace("_no_gas", "") + "_v3"

    if corresponding_name in dfs:
        new_length = len(dfs[corresponding_name])
        expected_length = original_lengths[corresponding_name] + count_added
        status = "✅ OK" if new_length == expected_length else f"❌ Mismatch (expected {expected_length}, got {new_length})"
        print(f"{corresponding_name}: added {count_added} copper rows — {status}")
```

```python
# Use pyproj's recommended Transformer
from pyproj import Transformer

# Transformer version of coordinate conversion
transformer = Transformer.from_crs("epsg:27700", "epsg:4326", always_xy=True)

def convert_coordinates(row):
    lon, lat = transformer.transform(row['EASTING'], row['NORTHING'])
    return pd.Series({'Latitude': lat, 'Longitude': lon})

# Update missing coordinates for copper rows only
for name, df in dfs.items():
    if {'SUBSTANCE NAME', 'EASTING', 'NORTHING'}.issubset(df.columns):
        # Make sure Latitude and Longitude columns exist
        if 'Latitude' not in df.columns:
            df['Latitude'] = pd.NA
        if 'Longitude' not in df.columns:
            df['Longitude'] = pd.NA

        mask = (df['SUBSTANCE NAME'].str.lower() == 'copper') & (
            df['Latitude'].isna() | df['Longitude'].isna()
        )

        if mask.any():
            df.loc[mask, ['Latitude', 'Longitude']] = df.loc[mask].apply(convert_coordinates, axis=1)
            dfs[name] = df  # update in-place
            print(f"✅ Updated coordinates for copper in {name}")
        else:
            print(f"ℹ️ No copper rows with missing coordinates in {name}")
    else:
        print(f"⚠️ Skipped {name} (missing required columns)")
```

```python
import json
from shapely.geometry import shape, Point

# Load the GeoJSON file
with open("Regions_December_2024_Boundaries_EN_BFC_1195854647342073399 (1).geojson") as f:
    regions_geojson = json.load(f)

# Extract region polygons and names
region_shapes = []
for feature in regions_geojson["features"]:
    region_name = feature["properties"]["RGN24NM"]
    polygon = shape(feature["geometry"])
    region_shapes.append((region_name, polygon))

# Function to find which region a point belongs to
def get_region_for_point(lat, lon):
    point = Point(lon, lat)
    for name, polygon in region_shapes:
        if polygon.contains(point):
            return name
    return None

# Assign region only to copper rows with coordinates
for name, df in dfs.items():
    if {'SUBSTANCE NAME', 'Latitude', 'Longitude'}.issubset(df.columns):
        # Ensure numeric types
        df['Latitude'] = pd.to_numeric(df['Latitude'], errors='coerce')
        df['Longitude'] = pd.to_numeric(df['Longitude'], errors='coerce')

        # Identify copper rows with missing REGION but valid coordinates
        mask = (df['SUBSTANCE NAME'].str.lower() == 'copper') & (
            df['Latitude'].notna() & df['Longitude'].notna()
        )

        if mask.any():
            df.loc[mask, 'REGION'] = df.loc[mask].apply(
                lambda row: get_region_for_point(row['Latitude'], row['Longitude']),
                axis=1
            )
            dfs[name] = df  # update in-place
            print(f"✅ REGION assigned for copper rows in {name}")
        else:
            print(f"ℹ️ No copper rows with valid coordinates in {name}")
    else:
        print(f"⚠️ Skipped {name} (missing required columns)")

```

```python
import json
from shapely.geometry import shape, Point
import pandas as pd

# Load the county GeoJSON file (geometry assumed to be in British National Grid: EPSG:27700)
with open("Counties_and_Unitary_Authorities_December_2023_Boundaries_UK_BFC_-3277701402845556292.geojson") as f:
    counties_geojson = json.load(f)

# Extract county polygons and names
county_shapes = []
for feature in counties_geojson["features"]:
    county_name = feature["properties"]["CTYUA23NM"]
    polygon = shape(feature["geometry"])
    county_shapes.append((county_name, polygon))

# Function to find county using Easting/Northing directly
def get_county_for_easting_northing(easting, northing):
    point = Point(easting, northing)
    for name, polygon in county_shapes:
        if polygon.contains(point):
            return name
    return None

# Assign counties to all valid rows in each DataFrame
for name, df in dfs.items():
    if {'EASTING', 'NORTHING'}.issubset(df.columns):
        # Ensure numeric types
        df['EASTING'] = pd.to_numeric(df['EASTING'], errors='coerce')
        df['NORTHING'] = pd.to_numeric(df['NORTHING'], errors='coerce')

        # Mask rows with valid coordinates
        mask = df['EASTING'].notna() & df['NORTHING'].notna()

        if mask.any():
            df.loc[mask, 'COUNTY'] = df.loc[mask].apply(
                lambda row: get_county_for_easting_northing(row['EASTING'], row['NORTHING']),
                axis=1
            )
            dfs[name] = df  # update in place
            print(f"✅ COUNTY assigned using Easting/Northing for {name}")
        else:
            print(f"ℹ️ No valid coordinate rows in {name}")
    else:
        print(f"⚠️ Skipped {name} (missing 'EASTING' or 'NORTHING')")
```
