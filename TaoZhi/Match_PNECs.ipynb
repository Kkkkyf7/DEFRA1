```python
import pandas as pd
from fuzzywuzzy import process

# Read csv file
csv_path = "C:/Users/ZITIAO/Desktop/NORMAN.csv"
df = pd.read_csv(csv_path, encoding='latin1')

# Organize a list of chemical names
chem_names = [
    "Polychlorinated biphenyls (PCBs)", "Mercury", "Nickel", "Cyanides",
    "Zinc", "Fluorides", "Phosphorus", "Cadmium", "Arsenic", "Chlorides",
    "Copper", "Indeno(1,2,3-cd)pyrene", "Phenols", "Anthracene", "Fluoranthene",
    "Vinyl chloride", "Hexachlorobutadiene", "Chromium", "Lead", "Naphthalene",
    "Benzene", "Toluene", "Xylene", "Di(2-ethylhexyl)phthalate (DEHP)", "Organotin compounds",
    "Tetrachloroethylene (PER)", "Dioxins and furans (PCDDs/PCDFs)", "Ethyl benzene",
    "Pentachlorophenol (PCP)", "Pentachlorobenzene", "Triphenyltin and compounds", "Atrazine",
    "Tributyltin and compounds", "Benzo(k)fluoranthene", "Octylphenols and octylphenol ethoxylates",
    "Hexachlorobenzene (HCB)", "Nonylphenols and nonylphenol ethoxylates",
    "Carbon tetrachloride (Tetrachloromethane)", "Diuron", "Trichloroethylene",
    "Chloroform (Trichloromethane)", "Ethylene dichloride (1,2-Dichloroethane)",
    "Dichloromethane (DCM)", "Benzo(a)pyrene", "Benzo(b)fluoranthene",
    "Ethylene oxide (1,2 Epoxyethane)", "Trifluralin", "Chlorfenvinphos",
    "Dichlorodiphenyltrichloroethane (DDT)", "Endosulfan", "Trichlorobenzene",
    "Cypermethrin", "Isodrin", "Aldrin", "Dichlorvos", "Dieldrin", "Endrin",
    "Simazine", "Hexachlorocyclohexane (HCH)", "Alachlor", "Benzo(g,h,i)perylene",
    "Chlordane", "Heptachlor", "Lindane", "Isoproturon", "Chlorpyrifos",
    "Asbestos", "Chlordecone", "Toxaphene", "Hexabromobiphenyl", "Mirex",
    "Perfluoro octyl sulphate (PFOS)", "Hexabromocyclododecane",
    "Brominated diphenylethers", "Short chain (C10-13) chlorinated paraffins (SCCPs)"
]

# Match chemical name
matched = []
for chem in chem_names:
    result = process.extractOne(chem, df['Substance'])
    if result:
        best_match, score = result[0], result[1]
        if score > 90:
            row = df[df['Substance'] == best_match].iloc[0]
            matched.append({
                "ChemName_in_Image": chem,
                "ChemName_in_Table": best_match,
                "PNEC": row['Lowest PNEC Freshwater [??g//l]'],
                "Date": row['Date']
            })


result_df = pd.DataFrame(matched)
display(result_df[['ChemName_in_Image', 'ChemName_in_Table', 'PNEC', 'Date']])
```
